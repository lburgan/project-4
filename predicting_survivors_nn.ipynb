```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf
```

```python
# Import our input dataset
titanic_df = pd.read_csv('titanic.csv')
titanic_df.head()
```

```python
#drop unecessary columns
titanic_df = titanic_df.drop(["name","ticket","cabin","boat","body","home.dest"], axis = 1)
```

```python
#get rid of missing values
new_df = titanic_df.dropna()
```

```python
new_df["pclass"] = new_df["pclass"].astype("int")
new_df["survived"] = new_df["survived"].astype("int")
```

```python
#get dummy varibles for Sex and Embarkment columns
dummy_df = pd.get_dummies(new_df)
dummy_df.head()
```

```python
# Remove survived target from features data
y = dummy_df.survived.values
X = dummy_df.drop(columns="survived").values

# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1912, stratify=y)
```

```python
# Preprocess numerical data for neural network

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the deep learning model
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=100, activation="sigmoid", input_dim=10))
nn_model.add(tf.keras.layers.Dense(units=50, activation="sigmoid"))
nn_model.add(tf.keras.layers.Dense(units=25, activation="sigmoid"))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=200)

# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
